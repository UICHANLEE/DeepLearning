```python
import pandas as pd
df = pd.read_csv('Heart.csv')
df
```

## 전처리 : 결측치 제거

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

## 전처리 : X/y 데이터 나누기

```python
y = df.pop('AHD')
X = df
```

```python
pd.get_dummies(y)
```

```python
# y 값의 인코딩
y[y=='Yes'] = 1
y[y=='No'] = 0
y
```

```python
# x값의 인코딩
pd.get_dummies(X)
```

## 전처리 : X값의 정규

```python
X = (X-X.mean())/X.std()
```

```python
X
```

```python
X_std = X
X_std
```

```python
import numpy as np
X.std = np.array(X_std, dtype = float)
y = np.array(y, dtype = float)
```

## 모델 만들기

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
def build_model():
  model = tf.keras.Sequential([
                               layers.Dense(64, input_shape=[X_std.shape[1]], activation = 'relu'),
                               layers.Dense(64, activation = 'relu'),
                               layers.Dense(1, activation = 'softmax') # 0과 1사이의 확률을 같도록                      
  ])
  model.compile(loss= 'binary_crossentropy', optimizer= 'adam', metrics = ['accuracy'])
  return model
```

```python
model = build_model()
model.summary()
```

## 모델학

```python
history = model.fit(X_std, epochs= 1000, validation_split = 0.2)
```
